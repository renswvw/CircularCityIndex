```python
# read the raw excel file from the url
excelFile = pd.ExcelFile(cwd + '/raw/D3_rawdata_27102022.xlsx')
df = pd.read_excel(excelFile, 'Municipio')
df
```

```python
# List headers of columns to check raw data set
list(df.columns)
```

```python
# filter columns that refer to 'cobertura 30Mpbs' (see leyenda)
df = df[[
 'Comunidad Autónoma',
 'Provincia',
 'Municipio',
 'Cob. 30Mbps\n(junio 2021)']]

# create a dictionary
# key = old name
# value = new name
dict = {'Comunidad Autónoma': 'Autonomous_Community',
        'Provincia': 'Province',
        'Municipio': 'Municipality',
        'Cob. 30Mbps\n(junio 2021)': '%_People_with_Broadband_Connection'}
 
# call rename () method
df.rename(columns=dict,
          inplace=True)

df
```

```python
# Split municipality column and combine to reorder names
df2 = pd.concat([df['Municipality'], df['Municipality'].str.split(', ', expand=True)], axis=1) ## split municipality column by comma to split article from municipality name (e.g., Rioja, La --> Rioja & La) 
df2['Municipality_clean'] = df2[1] + ' ' + df2[0] ## place article before municipality name (e.g., Rioja, La --> La Rioja) 
df2['Municipality_clean'] = df2['Municipality_clean'].fillna(df2.pop('Municipality')) 

# Filter out errors in municipality names
df['Municipality'] = df2['Municipality_clean'].str.replace("' ", "'") ## remove space between article and word when using an apostrophe before municipality name (e.g., L' Arbo --> L'Arbo')

# MORE NEEDED HERE
df
```

```python
# Title case to all column names
df['Municipality'] = df['Municipality'].str.title()
```

```python
# exports the dataframe into csv file with
folder = 'interim'
file_name = 'D3_interimdata'
date = datetime.now().strftime("%d%m%Y")
data_format = '.csv'

export_name = '/' + folder + '/' + file_name + '_' + date + data_format

df.to_csv(cwd + export_name, index=False)
```

```python
# exports the dataframe into excel file with
folder = 'interim'
file_name = 'D3_interimdata' 
date = datetime.now().strftime("%d%m%Y")
data_format = '.xlsx'

export_name = '/' + folder + '/' + file_name + '_' + date + data_format

df.to_excel(cwd + export_name, index=False)
```
