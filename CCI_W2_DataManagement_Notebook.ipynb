```python
# importing the required modules
from IPython.display import display
import glob
import pandas as pd
from pathlib import Path
import os
import re
from os import path
from datetime import datetime 
```

```python
# Get the current working directory
cwd = os.getcwd()

# Print the current working directory
print("Current working directory: {0}".format(cwd))
```

```python
# specifying the path to excel files
path = cwd + '/raw'
#path = 'C:\Users\rensw\OneDrive - Universiteit Utrecht\GIMA\Thesis - Buitenland\Thesis\Data\Raw_data\W2/raw'
print(path)
```

```python
# excel files in the path
file_list = glob.glob(path + "/*.xlsx")

# list of excel files we want to merge.
#pd.read_excel(file_path) #reads the excel
# data into pandas dataframe.
excl_list = []
```

```python
for file in file_list:  
    # get name of autonomous community
    file_name = os.path.basename(file)
    file_name_split = file_name.split("_")
    autonomousComm_name = file_name_split[1]
    autonomousComm_name = re.sub(r"(\w)([A-Z])", r"\1 \2", autonomousComm_name)

    # get name of province / consell comarcal   
    province_name = file_name_split[2]
    province_name = Path(province_name).stem
    province_name = re.sub(r"(\w)([A-Z])", r"\1 \2", province_name)
    
    # get name of province / consell comarcal  
    df = pd.read_excel(file)
    df['Autonomous_Community'] = autonomousComm_name
    df['Province'] = province_name

    excl_list.append(df)
```

```python
# create a new dataframe to store the
# merged excel file.
excl_merged = pd.DataFrame()
 
for excl_file in excl_list:
     
    # appends the data into the excl_merged
    # dataframe.
    excl_merged = excl_merged.append(
      excl_file, ignore_index=True)
  
# print original data frame
data_top = excl_merged.head()
display(data_top)
```

```python
# create a dictionary
# key = old name
# value = new name
dict = {'Año': 'Year',
        'Municipio': 'Municipality',
        'Nº de contenedores': 'Number of bins',
        'Envases de vidrio (kgs) recogidos': 'Glass packaging (kg) collected'}
 
# call rename () method
excl_merged.rename(columns=dict,
          inplace=True)

# print Data frame after rename columns
display(excl_merged)   
```

```python
#change order of columns
excl_merged = excl_merged.iloc[:, [0,4,5,1,2,3]]
excl_merged 
```

```python
# data cleaning of summed up rows per seperate table
# excl_merged[excl_merged['Municipality'].str.contains('TOTAL')==False]

excl_merged = excl_merged.drop(excl_merged.index[excl_merged['Municipality'] == "TOTAL"])
excl_merged
```

```python
# converting 'Year' from float to int
excl_merged['Year'] = excl_merged['Year'].astype(int)
  
# displaying the datatypes
display(excl_merged.dtypes)
print(excl_merged)
```
