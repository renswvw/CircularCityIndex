```python
# read the raw ECR2 excel file per sheet
excelFile = pd.ExcelFile(cwd + '/raw/ECR2_rawdata_13102022.xlsx')
df_adaption = pd.read_excel(excelFile, 'Commitment - Adaption')
df_2020 = pd.read_excel(excelFile, 'Commitment - 2020')
df_2030 = pd.read_excel(excelFile, 'Commitment - 2030')
df_2050 = pd.read_excel(excelFile, 'Commitment - 2050')
df_2020AND2030 = pd.read_excel(excelFile, 'Commitment - 2020 and 2030')
df_2020OR2030 = pd.read_excel(excelFile, 'Commitment - 2020 or 2030')
```

```python
# filter to municipality name
df_adaption = df_adaption[['Signatories']]
df_2020 = df_2020[['Signatories']]
df_2030 = df_2030[['Signatories']]
df_2050 = df_2050[['Signatories']]
df_2020AND2030 = df_2020AND2030[['Signatories']]
df_2020OR2030 = df_2020OR2030[['Signatories']]

df_adaption
```

```python
# change values to title-format
df_adaption['Signatories'].str.title()
df_2020['Signatories'].str.title()
df_2030['Signatories'].str.title()
df_2050['Signatories'].str.title()
df_2020AND2030['Signatories'].str.title()
df_2020OR2030['Signatories'].str.title()
```

```python
check_adaption = df_adaption['Signatories']
check_2020 = df_2020['Signatories']
check_2030 = df_2030['Signatories']
check_2050 = df_2050['Signatories']
check_2020AND2030 = df_2020AND2030['Signatories']
check_2020OR2030 = df_2020OR2030['Signatories']

df['Commitment_Adaption'] = df['Municipalities_Signed_CovenantofMayors'].apply(lambda str: any([(reqWord in str) for reqWord in check_adaption]))
df['Commitment_2020'] = df['Municipalities_Signed_CovenantofMayors'].apply(lambda str: any([(reqWord in str) for reqWord in check_2020]))
df['Commitment_2030'] = df['Municipalities_Signed_CovenantofMayors'].apply(lambda str: any([(reqWord in str) for reqWord in check_2030]))
df['Commitment_2050'] = df['Municipalities_Signed_CovenantofMayors'].apply(lambda str: any([(reqWord in str) for reqWord in check_2050]))
df['Commitment_2020AND2030'] = df['Municipalities_Signed_CovenantofMayors'].apply(lambda str: any([(reqWord in str) for reqWord in check_2020AND2030]))
df['Commitment_2020OR2030'] = df['Municipalities_Signed_CovenantofMayors'].apply(lambda str: any([(reqWord in str) for reqWord in check_2020OR2030]))

df
```
